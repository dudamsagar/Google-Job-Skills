```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```
# The Companies involved

plt.rcParams['figure.figsize'] = (7, 9)
data['Company'].value_counts().plot.bar(color = 'cyan')
plt.title('The companies Involved for recruitment', fontsize = 20)
plt.xlabel('Names of companies', fontsize = 15)
plt.show()
```

```
# checking most popular top 20 types of job Titles 

plt.rcParams['figure.figsize'] = (19, 8)
data['Title'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'pink')
plt.title("Most Popular 20 Job Titles of Google", fontsize = 20)
plt.xlabel('Names of Job Titles', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```
# checking most popular top 20 types of job Titles 

plt.rcParams['figure.figsize'] = (19, 8)
data['Category'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightgreen')
plt.title("Most Popular 20 Job Categories of Google", fontsize = 20)
plt.xlabel('Names of Job Categories', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```
# checking most popular top 20 types of job Locations

plt.rcParams['figure.figsize'] = (19, 8)
data['Location'].value_counts().sort_values(ascending = False).head(20).plot.bar(color = 'lightblue')
plt.title("Most Popular 20 Job Locations of Google", fontsize = 20)
plt.xlabel('Names of Job Locations', fontsize = 15)
plt.ylabel('count', fontsize = 15)
plt.show()
```

```
# checking most popular job destinations

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'white').generate(str(data['Location']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Places with Most Jobs', fontsize = 30)
plt.show()
```

```
# checking the most popular Responsibilities

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'pink', max_words = 100).generate(str(data['Responsibilities']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Responsibilities', fontsize = 30)
plt.show()
```

```
# checking the most popular Minimum edu. requirements

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'lightgreen', max_words = 80).generate(str(data['Minimum Qualifications']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Educational Requirements', fontsize = 30)
plt.show()
```

```
# checking the most popular Minimum edu. requirements

from wordcloud import WordCloud
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
wordcloud = WordCloud(background_color = 'yellow', max_words = 80).generate(str(data['Preferred Qualifications']))

print(wordcloud)
plt.rcParams['figure.figsize'] = (12, 12)
plt.axis('off')
plt.imshow(wordcloud)
plt.title('Most Popular Preferred Requirements', fontsize = 30)
plt.show()
```
